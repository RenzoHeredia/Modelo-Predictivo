```python
from sklearn.impute import SimpleImputer
import pandas as pd

# Crear el imputador para la media
imputer = SimpleImputer(strategy='mean')

# Imputar los valores faltantes en la columna específica
X['RAT_PROV_SAL_TC_log'] = imputer.fit_transform(X[['RAT_PROV_SAL_TC_log']])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_recall_curve
from imblearn.over_sampling import SMOTE  # Para manejar desbalance
from sklearn.model_selection import GridSearchCV
import numpy as np

# Paso 1: Manejar el desbalance con SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Paso 2: Crear y ajustar el modelo XGBClassifier con parámetros mejorados
model = XGBClassifier(
    objective='binary:logistic',
    max_depth=5,  # Reduce la profundidad para evitar sobreajuste
    learning_rate=0.1,  # Reduce el learning rate para mejorar la convergencia
    n_estimators=200,  # Más árboles para mayor capacidad
    scale_pos_weight=len(y_train_resampled[y_train_resampled == 0]) / len(y_train_resampled[y_train_resampled == 1]),  # Manejar desbalance
    random_state=42
)

# Entrenar el modelo
model.fit(X_train_resampled, y_train_resampled)

# Predecir probabilidades
y_proba = model.predict_proba(X_test)[:, 1]

# Paso 3: Ajustar el umbral de decisión
optimal_threshold = 0.3  # Ajusta este valor después de analizar las métricas
y_pred = (y_proba >= optimal_threshold).astype(int)

# Paso 4: Evaluar el modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')

print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))

print('Classification Report:')
print(classification_report(y_test, y_pred))

# Paso 5: Curva de precisión-recall para encontrar el mejor umbral (opcional)
precision, recall, thresholds = precision_recall_curve(y_test, y_proba)

# Encontrar el umbral que maximiza el F1-score
f1_scores = 2 * (precision * recall) / (precision + recall)
optimal_idx = np.argmax(f1_scores)
optimal_threshold = thresholds[optimal_idx]
print(f'Mejor umbral para F1-score: {optimal_threshold:.2f}')
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_recall_curve
from imblearn.over_sampling import SMOTE  # Para manejar desbalance
from sklearn.model_selection import train_test_split
import numpy as np

# Paso 1: Manejar el desbalance con SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Paso 2: Crear y ajustar el modelo XGBClassifier con parámetros mejorados
model = XGBClassifier(
    objective='binary:logistic',
    max_depth=5,  # Reduce la profundidad para evitar sobreajuste
    learning_rate=0.1,  # Reduce el learning rate para mejorar la convergencia
    n_estimators=200,  # Más árboles para mayor capacidad
    scale_pos_weight=len(y_train_resampled[y_train_resampled == 0]) / len(y_train_resampled[y_train_resampled == 1]),  # Manejar desbalance
    random_state=42
)

# Entrenar el modelo
model.fit(X_train_resampled, y_train_resampled)

# Predecir probabilidades
y_proba = model.predict_proba(X_test)[:, 1]

# Paso 3: Ajustar el umbral de decisión
optimal_threshold = 0.3  # Ajusta este valor después de analizar las métricas
y_pred = (y_proba >= optimal_threshold).astype(int)

# Paso 4: Evaluar el modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')

print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))

print('Classification Report:')
print(classification_report(y_test, y_pred))

# Paso 5: Curva de precisión-recall para encontrar el mejor umbral
precision, recall, thresholds = precision_recall_curve(y_test, y_proba)

# Calcular el F1-score de manera segura
f1_scores = np.where((precision + recall) == 0, 0, 2 * (precision * recall) / (precision + recall))

# Encontrar el índice del umbral que maximiza el F1-score
optimal_idx = np.argmax(f1_scores)
optimal_threshold = thresholds[optimal_idx]
print(f'Mejor umbral para F1-score: {optimal_threshold:.2f}')

# Paso 6: Recalcular predicciones con el mejor umbral
y_pred_optimal = (y_proba >= optimal_threshold).astype(int)

# Evaluar el modelo con el nuevo umbral
accuracy_optimal = accuracy_score(y_test, y_pred_optimal)
print(f'Accuracy con mejor umbral: {accuracy_optimal:.4f}')

print('Confusion Matrix con mejor umbral:')
print(confusion_matrix(y_test, y_pred_optimal))

print('Classification Report con mejor umbral:')
print(classification_report(y_test, y_pred_optimal))
```
