```python
X=df
X=X.drop(['TARGET'],axis=1)
X
```

```python
Y=df["TARGET"]
Y
```

```python
from sklearn.impute import SimpleImputer
import pandas as pd

# Crear el imputador para la media
imputer = SimpleImputer(strategy='mean')

# Imputar los valores faltantes en la columna específica
X['RAT_PROV_SAL_TC_log'] = imputer.fit_transform(X[['RAT_PROV_SAL_TC_log']])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from xgboost import XGBClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

#modelo = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
modelo = XGBClassifier(
    objective='binary:logistic',
    max_depth=6,# produnfidad de los arboles
    learning_rate=0.3, #tasa de aprendizaje 
    n_estimators=30000,# nuermo de arboles
    random_state=42
)
modelo.fit(X_train, y_train)

# Obtener las probabilidades para la clase positiva (1)
y_scores = modelo.predict_proba(X_test)[:, 1]

# Calcular FPR, TPR y AUC
fpr, tpr, thresholds = roc_curve(y_test, y_scores)
roc_auc = roc_auc_score(y_test, y_scores)

# Graficar la curva ROC
plt.figure()
plt.plot(fpr, tpr, color='blue', label=f"ROC curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], color='red', linestyle='--')  # Línea diagonal de referencia
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
# Definir el modelo GBDT (Gradient Boosting Decision Trees)
modelo = GradientBoostingClassifier(
    n_estimators=1,  # Número de árboles
    max_depth=6,       # Profundidad de los árboles
    learning_rate=0.3, # Tasa de aprendizaje
    random_state=42
)

# Entrenar el modelo
modelo.fit(X_train, y_train)

# Obtener las probabilidades para la clase positiva (1)
y_scores = modelo.predict_proba(X_test)[:, 1]

# Calcular FPR, TPR y AUC
fpr, tpr, thresholds = roc_curve(y_test, y_scores)
roc_auc = roc_auc_score(y_test, y_scores)

# Graficar la curva ROC
plt.figure()
plt.plot(fpr, tpr, color='blue', label=f"ROC curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], color='red', linestyle='--')  # Línea diagonal de referencia
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Crear el modelo GradientBoostingClassifier (GBDT)
model = GradientBoostingClassifier(
    n_estimators=100,  # Número de árboles
    max_depth=6,        # Profundidad de los árboles
    learning_rate=0.3,  # Tasa de aprendizaje
    random_state=42
)

# Entrenar el modelo
model.fit(X_train, y_train)

# Realizar las predicciones
y_pred = model.predict(X_test)

# Calcular la precisión y otras métricas
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')

# Mostrar la matriz de confusión
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))

# Mostrar el reporte de clasificación
print('Classification Report:')
print(classification_report(y_test, y_pred))
```
