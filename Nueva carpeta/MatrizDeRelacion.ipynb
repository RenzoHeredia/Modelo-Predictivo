```python
import lightgbm as lgb
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Crear el modelo LightGBM
model_lgb = lgb.LGBMClassifier(
    objective='binary',         # Para clasificación binaria
    max_depth=6,                # Profundidad máxima del árbol
    learning_rate=0.3,          # Tasa de aprendizaje
    n_estimators=100,           # Número de árboles (estimadores)
    random_state=42             # Fijar la semilla para reproducibilidad
)

# Entrenar el modelo
model_lgb.fit(X_train, y_train)

# Realizar las predicciones
y_pred = model_lgb.predict(X_test)

# Para LightGBM también, si estás usando clasificación binaria, las predicciones ya son 0 o 1
# No es necesario redondear, pero si quieres hacerlo puedes agregar esta línea:
y_pred = [1 if y > 0.5 else 0 for y in y_pred]

# Calcular la precisión y otras métricas
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')

# Matriz de confusión y reporte de clasificación
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))

print('Classification Report:')
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Crear el modelo RandomForest
model_rf = RandomForestClassifier(
    n_estimators=100,         # Número de árboles en el bosque
    max_depth=6,              # Profundidad máxima de los árboles
    random_state=42,          # Fijar la semilla para reproducibilidad
    n_jobs=-1,                # Usar todos los núcleos del procesador para acelerar el entrenamiento
    verbose=0                 # Controlar la salida de información durante el entrenamiento
)

# Entrenar el modelo
model_rf.fit(X_train, y_train)

# Realizar las predicciones
y_pred = model_rf.predict(X_test)

# Calcular la precisión y otras métricas
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')

# Matriz de confusión y reporte de clasificación
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))

print('Classification Report:')
print(classification_report(y_test, y_pred))
```
