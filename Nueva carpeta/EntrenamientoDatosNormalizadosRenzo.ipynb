```python
X=df_nor
X=X.drop(['TARGET'],axis=1)
X
```

```python
Y=df_nor["TARGET"]
Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Crear el modelo de regresión logística
model_lr = LogisticRegression(random_state=42, max_iter=1000)  # Aumentamos max_iter para garantizar convergencia

# Entrenar el modelo
model_lr.fit(X_train, y_train)

# Realizar las predicciones
y_pred_lr = model_lr.predict(X_test)

# Calcular la precisión
accuracy_lr = accuracy_score(y_test, y_pred_lr)
print(f'Accuracy (Logistic Regression): {accuracy_lr:.4f}')

# Matriz de confusión y reporte de clasificación
print('Confusion Matrix (Logistic Regression):')
print(confusion_matrix(y_test, y_pred_lr))

print('Classification Report (Logistic Regression):')
print(classification_report(y_test, y_pred_lr))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Definir el modelo de red neuronal
model_nn = Sequential([
    Dense(64, input_dim=X_train.shape[1], activation='relu'),  # Capa oculta con 64 neuronas y función de activación ReLU
    Dense(32, activation='relu'),                             # Otra capa oculta con 32 neuronas
    Dense(1, activation='sigmoid')                            # Capa de salida para clasificación binaria
])

# Compilar el modelo
model_nn.compile(optimizer=Adam(learning_rate=0.001),  # Optimizador Adam con tasa de aprendizaje de 0.001
                 loss='binary_crossentropy',           # Pérdida para clasificación binaria
                 metrics=['accuracy'])

# Entrenar el modelo
history = model_nn.fit(X_train, y_train, 
                       epochs=50,                         # Número de épocas
                       batch_size=64,                     # Tamaño de lote
                       validation_split=0.2,              # Dividir el 20% del entrenamiento para validación
                       verbose=1)                         # Mostrar el progreso del entrenamiento

# Realizar predicciones
y_pred_nn = (model_nn.predict(X_test) > 0.5).astype(int)  # Convertir probabilidades en 0 o 1

# Evaluar el modelo
accuracy_nn = accuracy_score(y_test, y_pred_nn)
print(f'Accuracy (Neural Network): {accuracy_nn:.4f}')

print('Confusion Matrix (Neural Network):')
print(confusion_matrix(y_test, y_pred_nn))

print('Classification Report (Neural Network):')
print(classification_report(y_test, y_pred_nn))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Definir el modelo SVM
# kernel='rbf' usa el kernel radial (uno de los más comunes), pero puedes probar otros como 'linear' o 'poly'
model_svm = SVC(kernel='rbf', C=1.0, gamma='scale', probability=True, random_state=42)

# Entrenar el modelo
model_svm.fit(X_train, y_train)

# Realizar las predicciones
y_pred_svm = model_svm.predict(X_test)

# Evaluar el modelo
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print(f'Accuracy (SVM): {accuracy_svm:.4f}')

# Matriz de confusión y reporte de clasificación
print('Confusion Matrix (SVM):')
print(confusion_matrix(y_test, y_pred_svm))

print('Classification Report (SVM):')
print(classification_report(y_test, y_pred_svm))
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Crear el modelo KNN
knn_model = KNeighborsClassifier(n_neighbors=5)  # Usamos 5 vecinos 

# Entrenar el modelo
knn_model.fit(X_train, y_train)

# Realizar las predicciones
y_pred_knn = knn_model.predict(X_test)

# Calcular la precisión
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(f'Accuracy (KNN): {accuracy_knn:.4f}')

# Matriz de confusión y reporte de clasificación
print('Confusion Matrix (KNN):')
print(confusion_matrix(y_test, y_pred_knn))

print('Classification Report (KNN):')
print(classification_report(y_test, y_pred_knn))
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler

# Escalar las características (recomendado para SGD)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Crear el modelo SGDClassifier
sgd_model = SGDClassifier(loss='hinge',  # Pérdida hinge para un clasificador SVM lineal
                          penalty='l2',  # Regularización L2
                          alpha=0.0001,  # Tasa de regularización
                          max_iter=5000,  # Número máximo de iteraciones
                          tol=1e-3,  # Tolerancia para detener el entrenamiento
                          random_state=42)  # Semilla para reproducibilidad

# Entrenar el modelo
sgd_model.fit(X_train_scaled, y_train)

# Realizar las predicciones
y_pred_sgd = sgd_model.predict(X_test_scaled)

# Calcular la precisión
accuracy_sgd = accuracy_score(y_test, y_pred_sgd)
print(f'Accuracy (SGDClassifier): {accuracy_sgd:.4f}')

# Matriz de confusión y reporte de clasificación
print('Confusion Matrix (SGDClassifier):')
print(confusion_matrix(y_test, y_pred_sgd))

print('Classification Report (SGDClassifier):')
print(classification_report(y_test, y_pred_sgd))
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Escalar las características
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Aplicar PCA
pca = PCA(n_components=0.95)  # Retener el 95% de la varianza explicada
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Crear el modelo SGDClassifier
sgd_model = SGDClassifier(loss='hinge',  # Pérdida hinge para un clasificador SVM lineal
                          penalty='l2',  # Regularización L2
                          alpha=0.0001,  # Tasa de regularización
                          max_iter=1000,  # Número máximo de iteraciones
                          tol=1e-3,  # Tolerancia para detener el entrenamiento
                          random_state=42)  # Semilla para reproducibilidad

# Entrenar el modelo
sgd_model.fit(X_train_pca, y_train)

# Realizar las predicciones
y_pred_sgd = sgd_model.predict(X_test_pca)

# Calcular la precisión
accuracy_sgd = accuracy_score(y_test, y_pred_sgd)
print(f'Accuracy (SGDClassifier con PCA): {accuracy_sgd:.4f}')

# Matriz de confusión y reporte de clasificación
print('Confusion Matrix (SGDClassifier con PCA):')
print(confusion_matrix(y_test, y_pred_sgd))

print('Classification Report (SGDClassifier con PCA):')
print(classification_report(y_test, y_pred_sgd))
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Crear los modelos base
model_lr = LogisticRegression(random_state=42, max_iter=1000)
model_xgb = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')

# Crear el Voting Classifier
ensemble_model = VotingClassifier(estimators=[
    ('logreg', model_lr),
    ('xgb', model_xgb)
], voting='hard')  # Cambia a 'soft' si quieres promediar las probabilidades

# Entrenar el modelo ensamble
ensemble_model.fit(X_train, y_train)

# Realizar predicciones
y_pred_ensemble = ensemble_model.predict(X_test)

# Calcular métricas para el modelo combinado
accuracy_ensemble = accuracy_score(y_test, y_pred_ensemble)
print(f'\nAccuracy (Ensemble - Logistic Regression + XGBoost): {accuracy_ensemble:.4f}')
print('Confusion Matrix (Ensemble):')
print(confusion_matrix(y_test, y_pred_ensemble))
print('Classification Report (Ensemble):')
print(classification_report(y_test, y_pred_ensemble))
```
